```python
import sys
 
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
            ]
directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
              }
 
 
def command_check(): # Функция которая узнает команду и запускает нужную функцию
   command = input('Введите команду: ')
   if command == 'p':
      p_command()
   elif command == 's':
      s_command()
   elif command == 'l':
      l_command()
   elif command == 'as':
      as_command()
   elif command == 'ds':
      ds_command()
   elif command == 'ad':
      ad_command()
   elif command == 'd':
      d_command()
   elif command == 'm':
      m_command()
   elif command == 'q':
      q_command()
   else: print('Неизвестная команда.')
      
def shelfs_list(): # Выводит текущий перечень полок
   print('Текущий перечень полок:', end=' ')
   for i in list(directories.keys()):
      print(i, end=' ')
   print()
 
def doc_num_check(x): # Проверяет, есть ли документ в базе по номеру документа
   if x in [j for i in list(directories.values()) for j in i]: # список всех номеров из directories
      return True      
   print('Документ не найден в базе')
   return False
 
def shelf_num_f(x): # Возвращает номер полки по номеру документа
   return [i for i in directories if x in directories[i]][0]
 
def shelf_num_check(y): # Проверяет, есть ли полка в базе по номеру полки
   if y in directories:
      return True
   print('Такой полки не существует')
   return False
 
def doc_request(): # Запрашивает номер документа
   doc_num = input('Введите номер документа: ')
   return doc_num
 
def shelf_request():# Запрашивает номер полки
   shelf_num = input('Введите номер полки: ')
   return shelf_num
 
def p_command():
   doc_num = doc_request()
   if doc_num_check(doc_num):
      for i in documents:
         if i['number'] == doc_num:
            print(f"Владелец документа: {i['name']}")
 
def s_command():
   doc_num = doc_request()
   if doc_num_check(doc_num):
      print(f"Документ хранится на полке: {shelf_num_f(doc_num)}")
 
def l_command():
   for i in documents:
      print(f"№: {i['number']}, тип: {i['type']}, владелец: {i['name']}, полка хранения: {shelf_num_f(i['number'])}")
 
def as_command():
   shelf_num = shelf_request()
   if not shelf_num_check(shelf_num):
      directories[shelf_num] = []
      print("Полка добавлена. "); shelfs_list()
   else:
      print('Такая полка уже существует. ', end=''); shelfs_list()
      
 
def ds_command():
   shelf_num = shelf_request()
   if not shelf_num_check(shelf_num):
      shelfs_list()
   elif shelf_num_check(shelf_num) and directories[shelf_num] == []:
      del directories[shelf_num]
      print("Полка удалена"); shelfs_list()
   else:
      print('На полке есть документы, удалите их перед удалением полки.', end=''); shelfs_list()
 
def ad_command():
   doc_num = doc_request()
   doc_type = input('Введите тип документа: ')
   doc_name = input('Введите владельца документа: ')
   shelf_num = shelf_request()
   if doc_num_check(doc_num):
      print('Такой номер документа уже есть в базе. База предполагает уникальность номеров для корректной работы команды d')
   elif not shelf_num_check(shelf_num):
      print('Добавьте полку командой as'); print('Текущий список документов: '); l_command()
   else:
      documents.append({'type': doc_type, 'number': doc_num, 'name': doc_name})
      directories[shelf_num].append(doc_num)
      print('Документ добавлен. ', end=''); print('Текущий список документов: '); l_command()
 
def d_command():
   doc_num = doc_request()
   if doc_num_check(doc_num):
      del directories[shelf_num_f(doc_num)][directories[shelf_num_f(doc_num)].index(doc_num)]
      for i in range(len(documents)):
         if documents[i]['number'] == doc_num:
            x = i
      del documents[x]
      print('Документ удален.')
   print('Текущий список документов: '); l_command()
 
def m_command():
   doc_num = doc_request()
   shelf_num = shelf_request()
   if not doc_num_check(doc_num):
      print('Текущий список документов: '); l_command()
   elif not shelf_num_check(shelf_num):
      shelfs_list()
   else:
      del directories[shelf_num_f(doc_num)][directories[shelf_num_f(doc_num)].index(doc_num)]
      directories[shelf_num].append(doc_num)
      print('Документ перемещен.')
      print('Текущий список документов: '); l_command()
 
def q_command():
   sys.exit(0)
 
print('''
                  Добрый день.
            Вот список команд:
      p - Находит владельца по номеру документа.
      s - Находит номер полки по номеру документа.
      l - Даёт всю информацию по документам.
      as - Позволяет добавить новую полку.
      ds - Позволяет удалить пустую полку.
      ad - Позволяет добавить новый документ.
      d - Позволяет удалить документ по номеру.
      m - Позволяет переместить документ на другую полку.
      q - Завершает программу.
      ''')
while True:
   command_check()
   print()
```
